```python
#Load CSV

import pandas as pd

df = pd.read_csv("../data/queens_trade_early_modern_1474_1816.csv")
df.head()
```

```python
# Add Reign Duration Column

df["reign_duration"] = df["reign_end"] - df["reign_start"]

df[["name", "reign_duration"]].sort_values("reign_duration", ascending=False)
```

```python
# Maritime vs Non-Maritime Analysis

df["maritime_power"].value_counts()
```

```python
# Add Visualization

import matplotlib.pyplot as plt

# Set style FIRST
plt.style.use("seaborn-v0_8-whitegrid")

earth_colors = [
    "#8B5A2B", "#A0522D", "#6B4226", "#B87333",
    "#7B3F00", "#9C661F", "#704214", "#C19A6B",
    "#8C7853", "#5C4033", "#A67B5B", "#996515",
    "#855E42", "#A97142", "#654321"
]

plt.figure(figsize=(12,6))

plt.bar(df["name"], df["reign_duration"], color=earth_colors[:len(df)])

plt.xticks(rotation=90)
plt.title("Reign Duration of Queens (1500–1700)", fontsize=14)
plt.ylabel("Years in Power")

plt.gca().set_facecolor("#F5F1E6")
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
